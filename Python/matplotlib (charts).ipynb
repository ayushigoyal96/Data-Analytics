```python
x=np.linspace(0,2*np.pi,360)
plt.plot([0,360],[0,0],'k')
sin=np.sin(x)
plt.plot(sin,ls='--')
plt.title('Sin')
plt.show()
```

```python
x=np.linspace(0,2*np.pi,360)
plt.plot([0,360],[0,0],'k')
cos=np.cos(x)
plt.plot(cos,ls='-')
plt.yticks(np.arange(-2,3))
plt.title('cos')
plt.show()
```

```python
x=np.linspace(0,2*np.pi,360)
l=list(x[:70]) + list(x[100:250]) + list(x[290:])
tan=np.tan(l)
plt.plot([0,360],[0,0],'k')
plt.plot(tan,ls='--')
plt.title('tan')
plt.show()
```

```python
fig=plt.figure(figsize=(15,5))
a1=fig.add_subplot(221)
x=np.linspace(0,2*np.pi,360)
l=list(x[:70]) + list(x[100:250]) + list(x[290:])
sin=np.sin(x)
cos=np.cos(x)
plt.plot(sin)
a2=fig.add_subplot(222)
plt.plot(cos)
plt.xticks(np.arange(0,0))
a3=fig.add_subplot(223)
tan=np.tan(l)
plt.plot(tan)
plt.show()
```

```python
fig=plt.figure(figsize=(15,8),edgecolor='blue')
a1=fig.add_subplot(111)
x=np.linspace(0,2*np.pi,360)
plt.plot([0,360],[0,0],'k',lw=3)
l=list(x[:70]) + list(x[100:260]) + list(x[290:])
sin=np.sin(x)
cos=np.cos(x)
tan=np.tan(l)
plt.plot(sin,'g',label='sin')
plt.plot(cos,'r',label='cos')
plt.plot(tan,'orange',label='tan')
plt.title("Trignometric Ratios",fontsize=30,color='blue')
plt.legend()
a2=fig.add_subplot(2,3,6)
a2.plot(sin,'g',label='sin')
plt.plot([0,360],[0,0],'k',lw=2)
plt.yticks(np.arange(-1,2))
plt.xticks(np.arange(0,0))
plt.legend()
plt.show()
```

```python
fig=plt.figure(figsize=(15,10))
x=np.linspace(0,2*np.pi,360)
l=list(x[:70]) + list(x[100:260]) + list(x[290:])
plt.plot([0,360],[0,0],'k',lw=3)
a1=fig.add_subplot(1,1,1)
sin=np.sin(x)
plt.plot(sin,'g',label='sin')
plt.yticks(np.arange(0,0))
plt.xticks(np.arange(0,0))
a2=fig.add_subplot(1,2,2)
cos=np.cos(x)
plt.plot(cos,'r',label='cos')
plt.xticks(np.arange(0,0))
plt.yticks(np.arange(0,0))
a3=fig.add_subplot(2,2,2)
tan=np.tan(l)
plt.plot(tan,'orange')
a4=fig.add_subplot(2,3,4)
plt.plot(sin,'g',label='sin')
plt.plot(cos,'r',label='cos')
plt.plot(tan,'orange',label='tan')
plt.legend()
plt.show()
```
