```python
plt.figure(figsize=(15,5))
x = np.linspace(0,2*np.pi,360)
l = list(x[:70]) + list(x[100:250]) +list(x[290:])
sin = np.sin(x)
cos = np.cos(x)
tan = np.tan(l)
plt.plot(sin,'r',lw=3,label="Sin")
plt.plot(cos,'g',lw=3,label = "Cos")
plt.plot(tan,'b',lw=3,label = "Tan")
plt.plot([0,360],[0,0],'k')
plt.title("Trignometric Ratios",fontsize=20,fontfamily="Comic Sans Ms",color="Red")
plt.xlabel("X - Axis",fontsize=20)
plt.ylabel("Y - Axis",fontsize=20)
#plt.text(25,-4,"Ravi",fontsize=20,rotation=45)
plt.xticks(np.arange(0,360,25))
plt.legend()
plt.grid(color='red',ls="-.")
plt.show()
```
