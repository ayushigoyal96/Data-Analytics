```python
bank = {'acc':[1001,1002,1003,1004],
        'user':['Ayushi','Anshul','Kartik','Naman'],
        'pass':['Python@123','abc@123','hello@123','sastenashe@123'],
        'bal':[12434324,325325435,463453242,342354365345235245]}
```

```python
from getpass import getpass
print("*"*125)
print("-"*125)
print("Welcome to Python Bank".center(125))
print("-"*125)
print("*"*125)
admin = 'root'
passwd = 'redhat'
while True:
    print("$ -- Main Menu -- $")
    print("\t1. Login\n\t2. SignUp\n\t3. Admin\n\t4. Exit")
    ch = input("\nEnter Your Choice --> ")
    if ch=="1":
        l_acc = int(input("\nEnter your account number --> "))
        if l_acc in bank['acc']:
            index = bank['acc'].index(l_acc)
            l_pass = getpass("\nEnter Your Password --> ")
            if l_pass == bank['pass'][index]:
                while True:
                    print("$ -- Login Menu -- $")
                    print("\t1. Credit\n\t2. Debit\n\t3. Check Balance\n\t4. Change Password\n\t5. Logout")
                    l_ch = input("\nEnter Your Choice --> ")
                    if l_ch=="1":
                            amount=int(input("Enter the amount-->"))
                            bank['bal'][index]+=amount
                            print("Credit Balance is-->",bank['bal'][index])
                    if l_ch=="2":
                            amount=int(input("Enter the amount-->"))
                            if bank['bal'][index]>=amount:
                                    bank['bal'][index]-=amount
                                    print("Net balance is-->",bank['bal'][index])
                            else:
                                print("Insufficient Balance")
                    if l_ch=="3":
                            print("Balance is --->",bank['bal'][index])
                        
                    if l_ch=="4":
                        l_pass=getpass("\nEnter Your New Password-->")
                        n_pass=getpass("\nRe-Type Your New password-->")
                        if l_pass==n_pass:
                            bank['pass'][index]=n_pass
                            print("\nPassword is successfully changed.")
                        else:
                            print("\nPassword not matched!!!!")
                    if l_ch == "5":
                        l_ch1 = input("\nDo you really want to Exit??? (Y/N) --> ").lower()
                        if l_ch1 == "yes" or l_ch1 == "y":
                            print("\nYou Are Successfully Logged Out!!!!")
                            break
                        else:
                            continue
                   
            else:
                print("\nInvalid Account Password!!")
        else:
            print("\nInvalid Account Number")
    elif ch == "2":
        print("\n$--Welcome to SignUp Menu--$")
        n_acc = bank['acc'][-1]+1
        n_user = input("\nCreate a UserName --> ")
        n_pass = getpass("\nCreate a Password --> ")
        v_pass = input("\nConfirm Password --> ")
        n_bal = 0.00
        if n_pass == v_pass:
            bank['acc'].append(n_acc)
            bank['user'].append(n_user)
            bank['pass'].append(n_pass)
            bank['bal'].append(n_bal)
            print(f"\nYour Account has been Created Successfully\nNow you can login!!!\nYour Account No. is  --> {n_acc}")
        else:
            print("\nPassword Doesnot Matched!!!!")
    elif ch == "3":
        adm_usr = input("\nEnter Admin UserName --> ")
        if adm_usr == admin:
            adm_pass = getpass("\nEnter Admin Password --> ")
            if adm_pass ==  passwd:
                print("Welcome Admin!!\nYou can see details of all accounts using Account Number!!")
                i_acc = int(input("\nEnter an account number to see details --> "))
                if i_acc in bank['acc']:
                    i = bank['acc'].index(i_acc)
                    print(f"Account No --> {bank['acc'][i]}")
                    print(f"UserName --> {bank['user'][i]}")
                    print(f"Balance --> ₹ {bank['bal'][i]}")
                else:
                    print("\nUser Account Doesnot Exists!!!!")
            else:
                print("\nIncorrect Admin Password!!!")
        else:
            print("\nInvalid Admin User!!!")
    elif ch == "4":
        ch1 = input("\nDo you really want to Exit??? (Y/N) --> ").lower()
        if ch1 == "yes" or ch1 == "y":
            print("\nThankYou For using Services!!!!")
            break
        else:
            continue
    else:
        print("\nInvalid Choice!!!")
```
